#### **=> Travelling Salesmen Problem - TSP(GA)**

```python
######## Features ########
"""
1- Selection = Tournament.
2- Fitness = -distance.
3- Mutation = Swap.
4- Chromosome = the arrangement (order) of the three cities.
5- The start and the end are always Benha.
6- Goal = find the SHORTEST ROUTE THROUGH ALL CITIES.
7- The Route = [Benha, C1, C2, C3, Benha].
8- The Fit city1 > city2 = The Fit city2 > city1.
9- Parameters is: Population Size = 50, MP = 30%, CP = 82%, No. of Generations = 30000
10- Initialize Population → Calculate Fitness → Selection → Crossover → Mutation → Repeat
"""
```

```python
import random
class TSB:
    def __init__(self, PopSize, cp, mp, num_of_gen):
        self.__PopSize = PopSize
        self.__cp = cp
        self.__mp = mp
        self.__num_of_gen = num_of_gen
        self.__cities = ["Shebin", "Zagazig", "Mansoura"]
        self.__distances = {
            "Benha": {"Shebin":25, "Zagazig":42, "Mansoura":79},
            "Shebin": {"Benha":25, "Zagazig":59, "Mansoura":91},
            "Zagazig": {"Benha":42, "Shebin":59, "Mansoura":63},
            "Mansoura": {"Benha":79, "Shebin":91, "Zagazig":63}
        }

        self.__population = []
        for i in range(self.__PopSize):
            individual = self.__cities.copy()
            random.shuffle(individual)
            self.__population.append(individual)

    #_______________________________________________________________________#
    def __calc_fitness(self):
        fitness_values = []
        for individual in self.__population:
            dist = 0
            way = ["Benha"] + individual + ["Benha"]
            for city in range(len(way) - 1):
                c1 = way[city]
                c2 = way[city + 1]
                dist += self.__distances[c1][c2]
            fitness_values.append(-dist)
        return fitness_values
    #_______________________________________________________________________#

    def __selection(self, fitness_vals):

        selected_pop = []

        for _ in range(self.__PopSize):

            # Tournament of size 2
            i1 = random.randint(0, self.__PopSize-1)
            i2 = random.randint(0, self.__PopSize-1)

            if fitness_vals[i1] > fitness_vals[i2]:
                winner = self.__population[i1]
            else:
                winner = self.__population[i2]

            selected_pop.append(winner.copy())

        return selected_pop
    #_______________________________________________________________________#
    def __crossover(self, parent1, parent2, probability):

        r = random.random()
        if r < probability:

            size = len(parent1)
            point = random.randint(1, size-1)

            child1 = parent1[:point]
            child2 = parent2[:point]

            for city in parent2:
                if city not in child1:
                    child1.append(city)

            for city in parent1:
                if city not in child2:
                    child2.append(city)

        else:
            child1 = parent1.copy()
            child2 = parent2.copy()

        return child1, child2    
    #_______________________________________________________________________#
    def __mutation(self, individual, probability):

        ind = individual.copy()
        r = random.random()

        if r < probability:

            i = random.randint(0, len(ind)-1)
            j = random.randint(0, len(ind)-1)

            ind[i], ind[j] = ind[j], ind[i]  # swap mutation

        return ind
    #_______________________________________________________________________#
    
    def __crossover_mutation(self):

        new_population = []

        for index in range(0, self.__PopSize, 2):

            parent1 = self.__population[index]
            parent2 = self.__population[index+1]

            child1, child2 = self.__crossover(parent1, parent2, self.__cp)

            new_population.append(child1)
            new_population.append(child2)

        for index in range(self.__PopSize):
            new_population[index] = self.__mutation(new_population[index], self.__mp)

        self.__population = new_population.copy()
    #_______________________________________________________________________#
    def solution(self):

        best_fitness_overall = None
        best_individual = None

        for gen in range(self.__num_of_gen):

            # Calculate fitness
            fitness_values = self.__calc_fitness()

            # Get best individual in current generation
            index_of_best = fitness_values.index(max(fitness_values))
            best_fitness = fitness_values[index_of_best]

            # Update global best
            if best_fitness_overall is None or best_fitness > best_fitness_overall:

                best_fitness_overall = best_fitness
                best_individual = self.__population[index_of_best].copy()

                print(f"\rGeneration: {gen:06} Best Distance = {-best_fitness}", end="")

            # Selection
            self.__population = self.__selection(fitness_values)

            # Crossover + Mutation
            self.__crossover_mutation()

        print("\nFinished all generations.")
        return best_individual
    #_______________________________________________________________________#
    def display_route(self, sol):

        way = ["Benha"] + sol + ["Benha"]
        dist_ans = 0

        print("\nBest Route:")
        print(" → ".join(way))

        for i in range(len(way) - 1):
            dist_ans += self.__distances[way[i]][way[i+1]]

        print("Total Distance =", dist_ans, "KM")
```

```python
obj = TSB(PopSize=50, cp=0.82, mp=0.30, num_of_gen=30000)

sol = obj.solution()
print("\nBest Chromosome:", sol)
obj.display_route(sol)
```
